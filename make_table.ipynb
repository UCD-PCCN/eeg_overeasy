Makes output table (as seaborn heatmap) for power analysis report.

rows: numbers of subjects
columns: effect sizes

assumes power values are formatted according to Kurt's power analysis function.

```python
# Import packages
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import eeg_pwr
```

```python
# testing import of eeg_pwr, and functionalization of power analysis

nSubs = [1,10,20,50]
ES = [1, 2, 3, 5]
data_array = np.array([1,2,3,5])
```

```python
pwrs = eeg_pwr.power_analysis(data_array, nSubs, ES)
```

```python
pwrs
```

```python
np.mean(pwrs,0)
```

```python
def print_power_table(mean_array, nSubs, ES):

    # Pivot dataframe so that rows are different sample sizes and columns are different effect sizes
    # df_pivot = df.pivot('Sample Size', 'Effect Size', 'Power')

    # Draw a heatmap with the numeric values in each cell
    f, ax = plt.subplots(figsize=(9, 6))
    sns.heatmap(mean_array, annot=True, fmt="f", linewidths=.5, ax=ax, xticklabels=ES, yticklabels=nSubs)
    ax.set_title('Power at different N and ES ')
    plt.xlabel('Effect Sizes')
    plt.ylabel('Numbers of subjects')
```

```python
print_power_table(np.mean(pwrs,0), nSubs, ES)
```
