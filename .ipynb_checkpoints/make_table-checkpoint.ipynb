Makes output table (as seaborn heatmap) for power analysis report.

rows: effect sizes
columns: numbers of subjects

Assuming input data frame of power values is in long format.

```python
# Import packages
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import eeg_pwr
```

```python
# Create example input data

test_data =  {'Sample Size': [10, 10, 10, 10, 20, 20, 20, 20, 30, 30, 30, 30, 40, 40, 40, 40],
              'Effect Size': [.1, .3, .5, .7, .1, .3, .5, .7, .1, .3, .5, .7, .1, .3, .5, .7],
              'Power': np.random.rand(16)}

df = pd.DataFrame(data=test_data)
```

```python
def print_output_table(df):

    # Pivot dataframe so that rows are different sample sizes and columns are different effect sizes
    df_pivot = df.pivot('Sample Size', 'Effect Size', 'Power')

    # Draw a heatmap with the numeric values in each cell
    f, ax = plt.subplots(figsize=(9, 6))
    sns.heatmap(df_pivot, annot=True, fmt="f", linewidths=.5, ax=ax)
    ax.set_title('Power at different N and ES ')
```

```python
print_output_table(df)
```
